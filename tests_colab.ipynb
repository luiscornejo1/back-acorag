### ğŸ’¬ Sobre el Escenario de Chat RAG (Nuevo)

El sistema de chat implementa **Retrieval-Augmented Generation (RAG)**:

1. **BÃºsqueda SemÃ¡ntica**: Encuentra documentos relevantes usando embeddings
2. **Filtrado por Relevancia**: Solo usa documentos con score > 0.20
3. **ConstrucciÃ³n de Contexto**: Formatea documentos como contexto para LLM
4. **GeneraciÃ³n con LLM**: Groq (llama-3.3-70b) genera respuesta contextualizada
5. **Persistencia**: Guarda conversaciones en PostgreSQL para anÃ¡lisis

**Ventajas del RAG:**
- âœ… Respuestas basadas en documentos reales (no inventadas)
- âœ… Cita fuentes especÃ­ficas (trazabilidad)
- âœ… Maneja preguntas fuera de contexto apropiadamente
- âœ… Historial permite contexto multi-turno
